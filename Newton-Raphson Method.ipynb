# Newton Raphson Method 

The formula for Newton-Raphson method is given below, 

    x_(new) = x - (f(x))/(f^'(x))

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = 7
#2x^2-5x+3=0
iteration = 0

for i in range (100):
    x_new = x - ((2*x**2)-(5*x)+3)/((4*x-5))
    if abs(x_new-x)<=0.000001:
        break
    iteration = iteration + 1
    print(iteration)
    x = x_new
print (x)
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = 10

#x^2-cos^2(x)+3

x_values = []
iteration_s = []
iteration = 0
for i in range (10000):
    x_new = x - ((x**2)-np.cos(x)**2-4*x)/(2*x+(2*np.sin(x)*np.cos(x)-4))
    if abs(x_new-x)<0.0000001:
        break
    iteration+=1
    iteration_s.append(iteration)
    x_values.append(x)
    x = x_new
print("The root: ", x)
print("The iteration: ", iteration)
plt.plot(iteration_s, x_values)
plt.xlabel('Iteration')
plt.ylabel('X values')
plt.show()
```
