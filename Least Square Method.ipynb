# Least Square Method 
 
y = a +bx

a = ((mean_y*Sum(x_i)^2)-(mean_x*Sum(x_i)*(y_i))/((Sum(x_i)^2-(n*mean(x)^2)) 

b = (((Sum(x_i)*(y_i))-(mean_x*Sum(y_i)))/((Sum(x_i)^2)-(n*mean_x^2)))

```python
import numpy as np 

x = np.array([3,4,5,6,7,8])
y = np.array([0,7,17,26,35,45])

n = len(x)

sum_x = 0
sum_y = 0
x_n = 0
y_n = 0
xy_n = 0

for i in range(n):
    sum_x += x[i]
    sum_y += y[i]
    x_n += x[i]**2
    y_n += y[i]
    xy_n += (x[i]*y[i])

mean_x = sum_x/n
mean_y = sum_y/n
a = (mean_y*x_n-mean_x*xy_n)/(x_n-(n*mean_x**2))
b = (xy_n-mean_x*y_n)/(x_n-(n*mean_x**2))

print('The value of a is : ', a)
print('The value of b is : ', b)

x_array = np.linspace(0,100,1000)
y_values = a+b*x_array

import matplotlib.pyplot as plt

plt.plot(x_array,y_values)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('True')
plt.show()



```

```python
import numpy as np 

x = np.array([3,4,5,6,7,8])
y = np.array([0,7,17,26,35,45])
n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)
sum_x = (np.sum(x**2))
sum_y = np.sum(y)
sum_xy = np.sum(x*y)

a = (y_mean*sum_x-x_mean*sum_xy)/(sum_x-(n*x_mean**2))
b = (sum_xy-x_mean*sum_y)/(sum_x-(n*x_mean**2))

print("The value of a is: ", a)
print("The value of b is: ", b)

x_array = np.linspace(0,100,1000)
y_values = a+b*x_array

import matplotlib.pyplot as plt

plt.plot(x_array,y_values)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('True')
plt.show()
```

```python
x = np.array([3,4,5,6,7,8])
y = np.array([0,7,17,26,35,45])
m = len(x)
n = 2
A = np.zeros((n+1,n+1))
B = np.zeros(n+1)
a = np.zeros(n+1)

for row in range (n+1):
    for col in range (n+1):
        if row ==0 and col ==0:
            A[row,col] = m
            continue
        A[row, col]= np.sum(x**(row+col))
    B[row] = np.sum(x**row *y)

a = np.linalg.solve(A,B)

print ('The coefficients of the polynomials: ', a)

import matplotlib.pyplot as plt

x_p = np.linspace(3,8,1000)
y_p = a[0] + a[1]*x_p +a[2]*x_p**2

plt.plot(x_p,y_p)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('True')
plt.legend(['Polynomial fit','Array given'])
plt.show()
```
