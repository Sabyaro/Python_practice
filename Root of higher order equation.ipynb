```python
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
x = 0.2
x_values=[]
x_values_new=[]
itrr = []
iteration = 0
for i in range(10000):
    x_new = (x**2+1)/2
    if abs(x-x_new)<0.000001:
        break
    x = x_new
    x_values.append(x)
    x_values_new.append(x)
    iteration = iteration+1
    itrr.append(iteration)
print(x)
plt.plot(itrr,x_values)
plt.plot(itrr,x_values_new)
plt.xlabel('Iteration number')
plt.ylabel('X values')
plt.legend(['X', 'X New'])
plt.grid('True')
plt.show()
```

```python
import numpy as np 
from math import sqrt
import matplotlib.pyplot as plt

x_new = 2
x = 5
error = 1
iteration = 0
x_values = []
x_values_new = []
itera = []
while abs(x-x_new)>0.00001:
    x = x_new
    x_new = sqrt(2*x-1)
    x_values.append(x)
    x_values_new.append(x_new)
    iteration = iteration+1
    itera.append(iteration)
print(x)
plt.plot(itera, x_values)
plt.plot(itera,x_values_new)
plt.xlabel('iteration number')
plt.ylabel('x values')
plt.legend(['x','x_new'])
plt.grid('True')
plt.show()


```
