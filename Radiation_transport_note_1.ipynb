## Radiation Field Variables 

### Direction and Solid angle convention



```python
from PIL import Image
import matplotlib.pyplot as plt
im = Image.open('image_1.png')
plt.imshow(im)
```

The directional properties of radiation fields are almost universally described using spherical polar coordinates. The direction vector is a unit vector, given in terms of the orthogonal Cartesian unit vectors **i**,**j**, and **k** by
$$
\begin{align}
\vec{\Omega} = \boldsymbol{i}u+\boldsymbol{j}\upsilon+\boldsymbol{k}\omega = \boldsymbol{i}\sin{\vartheta}\cos{\psi}+\boldsymbol{j}\sin{\vartheta}\sin{\psi}+\boldsymbol{k}\cos{\vartheta} \label{eq:1.1}
\end{align}
$$

Increase in $\vartheta$ by d$\vartheta$ and $\psi$ by d$\psi$ sweeps aout area d$\boldsymbol{A} = \sin{\vartheta}d\vartheta d\psi$ on a unit sphere. The differential solid angle associated with differential area d$\boldsymbol{A}$ is: 
$$
d\Omega = \sin{\vartheta}d\vartheta d\psi.\label{eq:1.2}
$$

A substantial simplification in notation can be achieved by making use of $\omega = \cos{\vartheta}$ as an independent variable instead of the angle $\vartheta$, so that $\sin{\vartheta}d\vartheta \equiv -d\omega$. A derivative benefit is a more straightforward integration over direction and the avoidance of a common error. The benefit is evident when one computes the solid angle subtended by "all possible directions", namely, 
$$
\vec{\Omega} = \int_{0}^{\pi}d\vartheta\sin{\vartheta}\int_{0}^{2\pi}d{\psi}=\int_{-1}^{1}d\omega \int_{0}^{2\pi}d\psi = 4\pi \label{eq:1.3}
$$

### Radiation Fluence

Fundamental way to characterize radiation is in terms of number of particles entering a volume. To make this charaterization, the radiometric concept of *fluence* is introduced. The particle fluence, or simply fluence at any point in a radiation field may be thought of in terms of the number of particles $\Delta N_p$ that, during some period of time, penetrate a hypothetical sphere of cross section $\Delta \boldsymbol{A}$ centered on the point in the figure below. 


```python
from PIL import Image
import matplotlib.pyplot as plt
im = Image.open('image_2.png')
plt.imshow(im)
```

The fluence defined as: 
$$
\Phi \equiv \lim_{\Delta\boldsymbol{A} \to 0} \frac{\Delta N_p}{\Delta\boldsymbol{A}} = \frac{dN_p}{d\boldsymbol{A}}
\label{eq:1.4}
$$
$$
\Phi \equiv \lim_{\Delta V \to 0} \frac{\sum_{i} s_i}{\Delta V}
\label{eq:1.5}
$$
$\Delta N_p$ = Number of particles penetrating the sphere with cross sectional area. It is the expectation value of the number of particles entering the sphere. $\Delta \boldsymbol{A}$ and
$s_i$ = path-length segments. 
The fluence rate or *flux density*, is expressed in terms of the number of particles entering a sphere, or the sum of path segments traversed within a sphere, per unit time, namely, 
$$
\phi \equiv \frac{d\Phi}{dt} = \frac{d^2N_p}{d\boldsymbol{A}dt}
\label{eq:1.6}
$$



### Net Flow and Net Flow Rate
Another way to characterize Radiation Field (RF) is the net number of particles crrossing a surface with well defined orientation. 


```python
from PIL import Image
import matplotlib.pyplot as plt
im = Image.open('image_3.png')
plt.imshow(im)
```

In this text, *net particle flow* (or simply *net flow*) at a point on a surface is the net number of particles in some specified time interval that flow across a unit differential area on the surface, in the direction specified as positive. As shown in the figure, one side of the surface is characterized as the positive side and is identified by a unit vector $\boldsymbol{n}$ normal to the area $\Delta \boldsymbol{A}$. If the number of particles crossing $\delta \boldsymbol{A}$ from the negative to the positive side $\Delta M_{p}^{+}$ and the number from the positive to the negative side $\Delta M_{p}^{-}$, then the *net* number crossing toward the positive side is $\Delta M_{p} \equiv \Delta M_{p}^{+} - \Delta M_{p}^{-}$. The net flow at the given point is designated as $J_n$, with the subscript denoting the unit normal $\boldsymbol{n}$ from the surface, and is defined as: 
$$
J_n \equiv \lim_{\Delta \boldsymbol{A} \to 0} \frac{\Delta M_p}{\Delta  \boldsymbol{A}}=\frac{dM_p}{d\boldsymbol{A}}
\label{eq:1.7}
$$
The relation between the net flow and the positive and negative flows is: 
$$
J_n \equiv J_{n}^{+} - J_{n}^{-}
\label{eq:1.8}
$$
The net flow rate is expressed in terms of the net number of particles crossing an area perpendicular to unit vector $\boldsymbol{n}$ per unit area per unit time, namely,
$$
j_n \equiv \frac{dJ_n}{dt} = \frac{d^2 M_p}{dA dt}
\label{eq;1.9}
$$
Flow rates can also be defined, so that 
$$
j_n = j_{n}^{+}-j_{n}^{-}
$$
For the concept of the fluence, the area presented to incoming particles is independent of the direction of the particles, whereas for the particle flow concept, the orientation of the area is well defined.

### Directional Properties of the Radiation Field

#### Properties of the Fluencs

The nature of the particles is implicit, and the argument $\boldsymbol{r}$ in $\Phi(\boldsymbol{r})$ is sometimes implicit. With no other arguments, $\Phi$ or $\Phi(\boldsymbol{r}$, represents the total fluence, irrective of the particle energy or particle direction, that is integrated over all particle energies and directions. 
In many circumustances, it is necessary to broaden the scope of the fluence to include information about the energies and direactions of particles. To do so requires the use of *distribution functions*. For example, $\Phi(\boldsymbol{r},E) \equiv \lim_{\Delta E \to 0} \Phi(\boldsymbol{r},E) \Delta E$ is, at point $\boldsymbol{r}$, the *fluence energy spectrum* $-$ the fluence of paticles with energies betweem $E$ and $E+dE^2$.
The angular variable itself is the vector direction $\boldsymbol{\Omega}$. The direction is a function of the polar ad azimuthal angles $\vartheta$ and $\psi$. Similarly, the differential element of solid angle is a function of the same two variables, namely $d\Omega = \sin{\vartheta}d\vartheta d\psi  = d\omega d\psi$. Thus, $\boldsymbol{\Phi}(\boldsymbol{r},\boldsymbol{\Omega})$ or $\boldsymbol{\Phi}(\boldsymbol{r},\omega,\psi)d\omega d\psi$ is, at point $\boldsymbol{r}$, the angular fluence $-$ the fluence of particles with solid angles in $d\Omega$ about $\boldsymbol{\Omega}$. 
The joint energy and angular distibution of the fluence is defined in such a way that $\Phi(\boldsymbol{r}, E, \boldsymbol{\Omega})dEd\Omega$ is the fluence of particles with energies in dE about E and with directions about $d\Omega$ about $\boldsymbol{\Omega}$. These distribution functions are related by
$$
\boldsymbol{\Phi}(\boldsymbol{r}) = \int_{4\pi}d\Omega \boldsymbol{\Phi}(\boldsymbol{r},\boldsymbol{\Omega}) = \int_{0}^{\infty} dE \boldsymbol{\Phi}(\boldsymbol{r},E) = \int_{0}^{\infty}dE\int_{4\pi}d\Omega\boldsymbol{\Phi}(\boldsymbol{r},E,\boldsymbol{\Omega})
$$
Occasionally finding use in radiation shielding and dosimetry is a quantity called the *energy fluence*. It is given by,
$$
\boldsymbol{\Psi}(\boldsymbol{r}) = \int_{0}^{\infty}dE\int_{4\pi}d\Omega\boldsymbol{\Phi}(\boldsymbol{r},E,\boldsymbol{\Omega})
$$

### Angular Properties of the Flow and Flow Rate

#### Definition of the Angular Flow

It is perfectly proper to define the angular flow in such a way that $J_n(\boldsymbol{r},\boldsymbol{\Omega})d\Omega$ is the flow of particles through a unit area with directions in $d\omega$ about $\boldsymbol{\Omega}$. The angular flow rate is written as $j_n(\boldsymbol{r}, \boldsymbol{\Omega})$.

#### Angular Flow vs Angular Fluence

The figure given below illustrates particles within a differential element of direction $d\Omega$ about direction $\boldsymbol{\Omega}$ crossing a surface perpendicular to unit vector $\boldsymbol{n}$. It is apparent that if $\Delta \boldsymbol{A}$ is the cross sectional area of the sphere, then the corresponding area in the surface is $\Delta \boldsymbol{A} \sec{\vartheta}$, where $\cos{\vartheta}=\boldsymbol{n}.\boldsymbol{\Omega}$. Thus, since the same number of particles pass through the sphere and through the area in the surface, $J_n(\boldsymbol{r},\boldsymbol{\Omega})\Delta A = \cos{\vartheta}\Delta A\boldsymbol{\Phi}(\boldsymbol{r},\boldsymbol{\Omega})$, or
$$
J_n(\boldsymbol{r},\boldsymbol{\Omega})\Delta A = \boldsymbol{n}.\boldsymbol{\Omega} \boldsymbol{\Phi}(\boldsymbol{r},\boldsymbol{\Omega}).
$$
The net flow is given by
$$
J_n(\boldsymbol{r}) \equiv \int_{4\pi} d\Omega J_n(\boldsymbol{r},\boldsymbol{\Omega}) = \int_{4\pi} d\Omega \boldsymbol{n}.\boldsymbol{\Omega} \boldsymbol{\Phi}(\boldsymbol{r},\boldsymbol{\Omega}).
$$
The fluence is a positive quantity; however, $J_n(\boldsymbol{r},\boldsymbol{\Omega})$ is positive or negative as $\boldsymbol{n}.\boldsymbol{\Omega}$ is positive or negative.

## Characterization of Radiation Source

### General Considerations 

* Point Source $-$ Most Fundamental 
    A real source can be approximated as a point source provided:
    
 1. That the volume is sufficiently small, that is, with dimensions much smaller than the dimensions of the attenuating medium between source and detector
 2. That there is negligible interaction of radiation with the matter in the source volume. 

The most general characterization of a point source used in this text is in terms of energy and direction, so that $S_p(E, \boldsymbol{\Omega}) dE d\Omega$ is the number of particles emitted with energies in dE about E and in $d\Omega$ about $\boldsymbol{\Omega}$. If it is understood that one is interested in the dose rate or some other response rate, then $S_p$ is the number of particles emitted *per unit time*. Common practical units for $S_p(E, \boldsymbol{\Omega})$ are $MeV^{-1}sr^{-1}$ or $MeV^{-1}sr^{-1}s^{-1}$.

Most radiation sources trated in shielding practive are isotropic, so that source characterization requires only knowledge of $S_p(E)d(E)$, which is the number of particles emitted with energies in dE about E(per unit time), and has common practical units of $MeV^{-1}$ or $MeV^{-1}s^{-1}$. 

Radiation sources may be distributed along a line, over an area, or within a volume. Source characterization requires, in general, spatial and energy dependence, with $S_l(\boldsymbol{r}, E)dE, S_a(\boldsymbol{r}, E)dE, and S_v(\boldsymbol{r},E)dE$ representing, respectively the number of particles emitted in dE about E per unit length, per unit area, and per unit volume.

## Linear Interaction Coefficients: 

$$
\mu_{i} = \lim_{\Delta x \to 0} \frac{P_i(\Delta x)}{\Delta x}
$$
$P_i(\Delta x)$ $-$ Probability that particle undergoes i-type reaction in path $\Delta x$ in matter. 

$\mu_{i}$ $-$ Linear Interaction Coefficient $-$ Probability per unit path length of i-type interaction

$\mu_{a}$ $-$ Linear Absorption Coefficient

$\mu_{s}$ $-$ Linear Scattering Coefficient

$\sum_{i}$ $-$ Macroscpic Cross Section

Total interaction coefficients:

$\mu_{t} = \sum_{i} \mu_{i}(E)$



## Attenuation of uncollided radiation

$I^{0}(0) = I_{0}$ $-$ particles/cm normally incident on a thick slab 

$$
P(\Delta x) = \frac{I^{0}(x) - I^{0}(x+dx)}{I^{0}(x)}
$$
$P(\Delta x)$ $-$ Probavility of interactionalong path length difference x+$\Delta x$ 

By definition 

$$
\mu_{t} = \lim_{\Delta x \to 0} \frac{P(\Delta x)}{\Delta x} = \lim_{\Delta x \to 0} \frac{I^{0}(x) - I^{0}(x+dx)}{I^{0}(x)} \frac{1}{I^{0}(x)} = -\frac{dI^{0}(x)}{dx} \frac{1}{I^{0}(x)}
$$

Therefore, 

$$
\frac{dI^{0}(x)}{dx} = - \mu_{t}I^{0}(x)
$$ 

And we get, 

$$
I^{0}(x) = I^{0}(0)e^{-\mu_{t}x}
$$

Uncollided inderctly-ionizing radiation attenuates exponentially as it passes through medium. 



```python
## Consider a thick slab of 10 cm and the total attenuation coefficient is 0.596 cm^-1 And the initial Intensity of the beam is 

import matplotlib.pyplot as plt
import numpy as np

def beam_intensity (I_0, mu_t, x):
    I = I_0*(np.exp(-(mu_t*x)))
    return I

#I_0 number of initial incident particle
#mu_t total linear attenuation coefficient
#x thickness of the slab
x = np.linspace(0,10,100)
mu_t = 0.596
I_0 = 10000

I = beam_intensity(I_0, mu_t, x)

plt.plot(x,I)
plt.xlabel('Thickness (cm)')
plt.ylabel('Intensity (number of particles)')
plt.title('Intensity vs Thickness Curve')
```

Probability of interaction along the path length x

$$
P(x) = 1 - \frac{I^{0}(x)}{I^{0}(0)} = 1 - e^{-\mu_{t}x}
$$

$$
\overline{P(x)} = 1 - P(x) = e^{-\mu_{t}x}
$$

$\overline{P(x)}$ = Probability of not interacting 

as x $\rightarrow$ dx p(dx) $\rightarrow$ $\mu_{t}dx$

$\boldsymbol{Mean Free Path}$ = Average distance travelled before interaction

p(x)dx $\rightarrow$ Probability of first interaction in x to x+dx = {probability of not interaction upto x}.{probability of ineraction in dx} = $\overline{P(x)}$P(dx)= $\mu_{t}$ $e^{-\mu_{t}x} dx$

Therefore the mean free path $\overline{x}$
$$
\overline{x} = \frac{\int_{0}^{\infty}p(x)dx}{\int_{0}^{\infty}P(x)dx} =  \frac{\mu_{t} \int_{0}^{\infty} x e^{-\mu_{t}x} dx}{1} = \frac{1}{\mu_{t}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
def interaction_along_path_length(mu_t, x):
    P = 1- np.exp(-mu_t*x)
    return P
def not_interaction_along_path(mu_t, x):
    P_ = np.exp(-mu_t*x)
    return P_

x = np.linspace(0,10,100)
y1 = interaction_along_path_length(0.596, x)
y2 = not_interaction_along_path(0.596, x)
plt.plot(x,y1, label = 'Probability of interaction', color = 'blue')
plt.plot(x,y2, label = 'Probability of not interaction', color = 'green')
plt.xlabel('Thickness (cm)')
plt.ylabel('Probability of interaction / Probability of not interaction (number of particles)')
plt.title('Probability of interaction')
```

$$
\mu_{i} \propto N
$$

$$

$$

Here, 

$\rho$ is the density of the material

$N_a$ Avogadro Number

## Uncollided Flux Density from an Isotropic Point Source

### Point Source in Vacuum

In case of slab problem as it is mentioned for the intensity calculation. The flux is 
$$
\phi^{0}(x) = \phi^{0}(0) e^{-\mu_{t}x}
$$

Point source $-$ Most basic Neutron Source 
             $-$ Mathematical idealization
             $-$ Small differential volume source can be treated like points 

Most radiation are isotropically emitted 

$S_p$ $\rightarrow$ Point of isotropic source into an infinite vacuum. All particles move rapidly outward without interaction and because of the source isotropy, each unit area on an imaginary spherical shell of radius r has the same number of particles crossing it namely, $\frac{S_p}{4\pi r^{2}}$.

The definition of the fluence $\Phi^{0}$ of uncollided particles a distance r from the source is 
$$
\Phi^{0}(r) = \frac{S_{p}}{4 \pi r^2}

$$
D^{0}(r) = \frac{S_p \Re}{4 \pi r^2}

In a small detector with volume $\Delta V_{d}$, interaction rate $R_{d}$, 

$$
R_{d} = \mu_{d} \Delta V_{d} \frac{S_p}{4 \pi r^2}
$$

Here, 

$\mu_{d}$ $-$ kinematice interaction coefficient of detector material 

Note that the dose and fluence decreases as $1/r^{2}$ as the distance from the source is increased. The decreasing dose with increasing distance is something referred to as geometric attenuation.

### Point Source in a Homogeneous Attenuating Medium

Now consider the same point monoenergetic isotropic source embedded in an infinite homogeneous medium characterized by a total interaction coefficient $\mu$. As the source particles stream radially outward, some interact before they reach the imaginary sphere of radius r and do not contribute to the uncollided fluence. The number of source particles that travel at least a distance r without interaction is $S_p e^{\mu r}$ , so that the uncollided dose is 
$$
D^{0}(r) = \frac{S_p \Re}{4 \pi r^2} e^{-\mu r}
$$

The term $e^{-\mu r}$ is referred to as the material attenuation term to distinguish it from the $1/r^2$ geometric attenuation term. 

#### Point Source with a Shield

Now suppose that the only attenuating material separating the source and the detector is a slab of material with attenuation coefficient $\mu$ and thickness $t$ as shown in the above figure. In this case the probability that a source particle reaches the detector with interaction is $e^{\mu t}$, so that the uncollided dose is, 

$$
D^{0}(r) = \frac{S_p \Re}{4 \pi r^2}e^{-\mu t}
$$

This same result holds if the attenuating medium has any shape $\left[e.g., a spherical shell of thickness t as shown in the figure above\right]$ 

If the interposing shield is composed of a series of different materials such that an uncoliided particle mush penetrate a thickness $t_i$ of a materials with attenuation coefficient $\mu_i$ before reaching the detector, the uncollided dose is
$$
D^{0}(r) = \frac{S_p \Re (E)}{4 \pi r^2} exp(-\sum_{i} \mu_{i} t_{i})
$$

Here $\sum_{i} \mu_{i} t_{i}$ is the total number of mean free path lengths of attenuating material that an uncoliided particle must traverse without interaction, and $exp(-\sum_{i} \mu_{i} t_{i})$ is the probability that a source particle traverses this number of mean free path lengths without interaction. 

For Series of material the fluenc calculation would be: 

$$
\Phi^{0}(r) = \frac{Sp}{4 \pi r^2} exp(-\sum_{i}\mu_i t_{i})
$$

#### Point Source in a Heterogeneous Medium

In this case the probability that a source particle reaches the detector a distance r away is $e^{-l}$, where $l$ is the total number of mean-free-path lengths of material that a particle must traverse without interaction to reach the detector. This dimensionless distance can be written formally as:
$$
l = \int_{0}^{r} ds \mu(s)
$$

Where the integration is along the ray from source to detector. The uncollided dose then is given by
$$
D^{0}(r) = \frac{S_p \Re}{4 \pi r^2} exp(-\int_{0}^{r} ds \mu(s))
$$

#### Polyenergetic Point Source 

If the source emits particles at several discrete energies, the total uncollided dose is found simply by summing the uncollided doses for each monoenergetic group of source particles. Thus if a fraction $f_i$ of the $S_p$ source particles are emitted with energy $E_i$, the total uncollided dose is:
$$
D^{0}(r) = \sum_{i}\frac{S_p \Re(E_i) f_i}{4 \pi r^2} exp(-\int_{0}^{r} ds \mu(s, E_i))
$$

For a source emitting a continuum of particle energies such that $N(E)dE$ is the probability that a source particle is emitted with energy in dE about E, the total uncollided dose is: 

$$
D^{0}(r) = \int_{0}^{\infty}dE\frac{S_p \Re(E) N(E)}{4 \pi r^2} exp(-\int_{0}^{r} ds \mu(s, E))
$$

### Line Source 

A straight line source of length L emitting isotropically $S_l$ particles per unit length at energy E is depicted in the figure below. A detectro is positioned at point P, a distance h from the source along a perpendicular to one end of the line. Consider a segment of the line source between distnce x and x+dx measured from the bottom of the source. The source within this segment may be treated as an effective point isotropic source emitting $S_ldx$ particles which produces an uncollided dose at P of $dD^{0}$. To obtain the total dose at P from all segments of the line source, one then must sum, or rather integrate, $dD^{0}$ over all line segments.

#### Line Source in a Nonattenuating Medium

In the asence of material interaction, the differential uncollided dose produced by particles emitted in dx about x is, 
$$
dD^{0}(P) = \frac{1}{4 \pi} \frac{S_l \Re dx}{x^2 + h^2}
$$
and thus
$$
D^{0}(P) = \frac{S_l \Re}{4 \pi} \int_{0}^{L} \frac{dx}{x^2 + h^2}
$$

As can be see from the above figure, $x = h \tan{\theta}$ and $x^2 + h^2 = h^2 \sec^2{\theta}$. It then follows that $dx=h \sec^2{\theta}d\theta$, so that the total uncollided dose at P becomes:
$$
D^{0}(P) = \frac{S_l \Re}{4 \pi h}\int_{0}^{\theta_{0}} d\theta= \frac{S_l \Re \theta_{0}}{4 \pi h}
$$
The angle $\theta_{0}$ in this expression must be expressed in radius. For $L<<h$, $\theta_{0}\simeq L/h$, so that the result above becomes the point-source result of namely
$$
D^{0}(P) = \frac{S_{tot} \Re}{4 \pi h^2}
$$

Where $S_{tot} \equiv S_lL$ is the total source strength. 

#### Line Source in a Homogeneous Attenuating Medium

Now suppose that the source and receptor are present in a homogeneous mediumwith a total interaction coefficient $\mu$. Attenuation along the ray from x to P reduces the uncollided at P to, 
$$
dD^{0}(P) = \frac{1}{4 \pi} \frac{S_l \Re dx}{x^2 + h^2} exp(-\mu \sqrt{x^2 + h^2})
$$

Where $\Re$ and $\mu$ generally depend on the particle energy E. The total uncollided dose now is described by the integral 
$$
D^{0}(P) = \frac{S_l \Re}{4 \pi h}\int_{0}^{\theta_{0}} d\theta e^{-\mu h \sec{\theta}}
$$

This integral can not be evaluated analytically. However, it can be expressed in terms of the *Sievert integral* of the *secant integral*, defined as
$$
F(\theta, b) \equiv \int_{0}^{\theta} dx e^{-b \sec{x}}
$$

With this integral, which is frequently encountered in shielding analysis, the above equation can be expressed as
$$
D^{0}(P) = \frac{S_l \Re}{4 \pi h} F(\theta_{0}, \mu h)
$$

#### Line Source Behind a Slab Shield

Suppose now that the only material separating the line source and the receptor is a parallel slab or concentric cylinder-shell shield of thickness t and total attenuation coefficient $\mu_s$. For this case, the analysis above using only attenuating in the slab yields

$$
D^{0}(P) = \frac{S_l \Re}{4 \pi h} F(\theta_{0}, \mu_s t)
$$

If the shield is made up od layers of thickness $t_i$ and attenuation coefficients $\mu_{si}$, then $\mu_s t$ must be replaced by $\sum_{i}\mu_{si}t_i$, the total mean free path thickness of the shield.  

### Disk Source 

An annular strip of radius $\rho$, width $d\rho$, and area $dA = 2 \pi \rho d\rho$ is shown in the figure below.  The width $d\rho$ is infinitesimal and it is clear that all sources within the area dA are equidistant from the receptor point P. If $\S_a$ is the area source emission, with units of inverse area, then the strength of an effective point source equivalent to the sources between $\rho$ and $\rho+d\rho$ is just $S_a dA$. The distance from this effective point source to the receptor is $r = \sqrt{\rho^{2} + h^2}$.

#### Disk Source in a Nonattenuating Medium

Only geometric attenuation, given by the inverse square law, applies and uncollided dose at the receptor, accounting for tha entire disk source, is given by:
$$
D^{0}(P) = \frac{S_a \Re}{4 \pi} \int_{A} \frac{dA}{r^2} = \frac{S_a \Re}{2} \int_{0}^{a}d\rho \frac{\rho}{\rho^2 + h^2}
$$

This integral may be evaluated analytically. Since h is fixed. $\rho d\rho = rdr$. One may also note that, since $r = h \sec{\theta}$, $dr = h \sec{\theta}\tan{\theta}d\theta$. It follows that, 
$$
D^{0}(P) = \frac{S_a \Re}{2}\int_{h}^{h \sec{\theta_{0}}} dr r^{-1} = \frac{S_a \Re}{2}\int_{0}^{\theta_{0}}d\theta \tan{\theta} = \frac{S_a \Re}{2} ln({\sec{\theta_{0}}})= \frac{S_a \Re}{4}ln(1+\frac{a^2}{h^2})
$$
For a<<h, $ln(1+a^2/h^2) \approx a^2/h^2$, so that the above equation may be approximated as, 
$$
D^{0}(P) \approx \frac{S_a \Re a^2}{4 h^2}=\frac{S_{tot}\Re}{4 \pi h^2}
$$
Where $S_tot = S_a \pi a^2$, the total number of particles emitted by the disk. 

#### Disk Source in a Homogeneous Attenuating Medium

Suppose that the source and receptor are present in a homogeneous medium with tota interaction coefficient $\mu$. Exponential attenuation along path r must now be taken into account, therefore the equation becomes: 
$$
D^{0}(P) = \frac{S_a \Re}{2}\int_{h}^{h \sec{\theta_{0}}} dr r^{-1}e^{-\mu r}
$$
A change of independent variable from r to $x \equiv \mu r$ allows this equation to be written in a form which introduces another special mathematical function useful in shielding analysis: 
$$
D^{0}(P) = \frac{S_a \Re}{2} \int_{\mu h}^{\mu h \sec{\theta_{0}}} dx x^{-1} e^{-x}
$$
or, 
$$
D^{0}(P) = \frac{S_a \Re}{2}(\int_{\mu h}^{\infty} dx x^{-1} e^{-x} - \int_{\infty}^{\mu h \sec{\theta_{0}}} dx x^{-1} e^{-x})
$$
With the exponential integral function, the uncollided dose is given by, 
$$
D^{0}(P) = \frac{S_a \Re}{2} (E_1(\mu h)-E_1(\mu h \sec{\theta_{0}})
$$
Where E is the *exponential integral function* $E_n(b)$ can be defined as:
$$
E_n(b) \equiv b^{n-1}\int_{b}^{\infty} dx x^{-n} e^{-x} = \int_{1}^{\infty}dx x^{-n} e^{-bx} = \int_{0}^{1} dx x^{n-2}e^{-b/x}
$$

#### Disk Source with a Slab Shield

Suppose that between the disk source and the receptor, a slab shield of thickness t and linear interaction coefficient $\mu_s$ is placed parallel to the source and that the medium between source and receptor is otherwise nonattenuating. It is easy to show that in this case, the uncollided dose is, 
$$
D^{0}(P) = \frac{S_a \Re}{2} (E_1(\mu_st)-E_1(\mu_s t\sec{\theta_{0}}))
$$

### Spherical Surface Source 

Consider a spherical suface of radius r that emits $S_a$ particles per unit area. The sphere is composed of a homogeneous attenuating medium with attenuation coefficient $\mu$. A differential area dA on the surface, in terms of polar coordinates, equals $r^2d\psi d(\cos{\theta})$, where $\theta$ and $\psi$ are the polar and azimuthal angles, respectively with respect to the center of the sphere. The detector response to uncollided radiation at the center is given by
$$
D^{0} = \int_{A} \frac{S_a \Re e^{-\mu r}}{4 \pi r^2} = \frac{S_a \Re e^{-\mu r}}{4 \pi} \int_{0}^{2 \pi} d\psi \int_{-1}^{1} d(\cos{\theta}) = S_a \Re e^{-\mu r}
$$
If no attenuating medium exists, then $D^{0} = S_a \Re$

#### Infinite Slab Source 

Consider a detector on or above the surface of a homogeneous slab of thickness H and interaction coefficient $\mu$ containing a uniform volumetric radiation source that emits $S_v$ particles per unit volume. This situation is just a limiting case of a frustrum of a cone as the frustrum's angle $\theta_0$ goes to $\pi/2$. As $\cos{\theta_0} \rightarrow 0$, the uncollided dose above the slab source is seen to be 
$$
D^{0}(P) = \frac{S_v \Re}{2 \mu_s} (1- E_2(\mu_s H))
$$
If the slab thickness H becomes infinite, the uncollided dose above an infinite half space source is 
$$
D^{0}(P) = \frac{S_v \Re}{2 \mu_s}
$$
If a parallel slab shield of thickness t and interaction coefficient $\mu$ is placed between the detector and slab source gives $\cos{\theta_0} \rightarrow 0$,

$$
D^{0}(P) = \frac{S_v \Re}{2 \mu_s} (E_2(\mu t)- E_2(\mu_s H + \mu t))
$$

For an infinte half-space source reduces to 
$$
D^{0}(P) = \frac{S_v \Re}{2 \mu_s} E_2(\mu t)
$$



### Cylindrical Source in an Attenuating Medium

The source emission is uniform of strength $S_v$ in a homogeneous medium with linear interaction coefficient $\mu$. Height and radius are, respectively, H and R. The uncollided doses ar point $P_0$ at the center of the base and at $P_1$ on the perimeter of the base are computed as the integrals over the entire volume of the cylinder of contributions from sources within elements of volume $dV = \rho d\rho d\psi dz$ trated as effective point sources. From points $P_0$ and $P_1$, the distances r to the volume element dV are determined by the law of cosines and by the Pythagorean theorem, namely for point $P_0$
$$
r^2 = z^2 + \rho^2
$$
and for point $P_1$
$$
r^2 = z^2 + R^2 + \rho^2 - 2\rho R \cos{\psi}
$$
The uncollided doses, accounting for symmetry in$\psi$ for point $P_0$, are thus, 
$$
D^{0}(P_0) = \frac{S_v \Re}{2}\int_{0}^{R} d\rho \int_{0}^{H} dz \frac{e^{-\mu r}}{r^2}
$$
and 
$$
D^{0}(P_1) = \frac{S_v \Re}{2 \pi} \int_{0}^{\pi} d\psi \int_{0}^{R} d\rho \rho \int_{0}^{H} dz \frac{e^{-\mu r}}{r^2}
$$
