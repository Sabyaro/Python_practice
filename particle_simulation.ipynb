```python
import numpy as np
import matplotlib.pyplot as plt

pos_incident = np.array([1,0,0])
pos_target = np.array([2,42,6])
dt = 0.01
Energy = 1.6e-13
v_0 = np.array([10,0,0])
Z = 22

def force(pos_incident,pos_target):
    distance = np.sqrt(sum(np.square(pos_target-pos_incident)))
    r =(pos_target-pos_incident)

    F =(9e9*Z*(1.6e-19)**2)*r/(distance**3)

    return F
m = 9.11e-31
v=np.zeros(3)
i=0
x=[]
y=[]
z=[]
t=0
while t<100:
    v = v_0 + (force(pos_incident,pos_target)/m)*dt

    pos_incident =pos_incident+ v*dt
    v_0=v
    pos_1 = pos_incident
    x.append(pos_1[0])
    y.append(pos_1[1])
    z.append(pos_1[2])
    t=t+dt

    print("Velocity", v)
    print("Position", pos_incident)
    
print("Velocity of the particle: ", v)
print("Position of the particle: ", pos_incident)
from mpl_toolkits import mplot3d

fig=plt.figure()
ax=plt.axes(projection='3d')

ax.plot3D(x,y,z,'green')
ax.scatter(pos_target[0],pos_target[1],pos_target[2],color='red')
ax.scatter(pos_incident[0],pos_incident[1],pos_incident[2],color='blue')
plt.show()

    

```
